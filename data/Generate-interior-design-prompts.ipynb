```python
res = bard.get_answer('''generate JSON objects with other 10 such examples, but with for living room for interior design editing, but with just 1 feature change at a time (change, addition or removal), different from the previous answers. Respect the most succint format like this: json is on a single line in format {"input:", "edit:", "output:"}, with one json per line, without other text between jsons and without any other message
{"input": "An elegant dining room with a long table that seats eight, a statement chandelier, and plush velvet chairs.", "edit": "Replace the statement chandelier with a modern pendant light.", "output": "An elegant dining room with a long table that seats eight, a modern pendant light, and plush velvet chairs."}
{"input": "A welcoming entryway with a large mirror, a console table with fresh flowers, and a stylish rug.", "edit": "Replace the console table with a floating bench.", "output": "A welcoming entryway with a large mirror, a floating bench with fresh flowers, and a stylish rug."}
{"input": "An organized mudroom with hooks for coats, shelves for shoes, a bench for seating, and cubbies for storage.", "edit": "Add a designated pet area with a food bowl, water bowl, and bed.", "output": "An organized mudroom with hooks for coats, shelves for shoes, a bench for seating, cubbies for storage, and a designated pet area with a food bowl, water bowl, and bed."}''')['content']
print(res)
```

```python
def write_jsons(response, filename = "prompts.txt"):
    with open(filename, "a") as f:
        for line in response.split("\n"):
            if len(line):
                try:
                    json.loads(line)
                    f.write(line + "\n")
                except:
                    print("incorect line")
        f.write("\n")
```

```python
res = bard.get_answer('''generate another various and different type of living rooms set with 10 with short edit prompts (like add x, replace x with y, remove x) and with jsons on the same line. Make edits with 1 feature at a time (ex. wall, chair, sink, table etc) and remove the numbering on each line''')['content']    
print(res)
```

```python
while True:
    res = bard.get_answer('''generate another various and different type of living rooms set with 10 without repeting the previous ones and with succint edit prompts (like add x, replace x with y, remove x) and no information regarding the feeling and with jsons on the same line. Make edits with 1 feature at a time (ex. wall, chair, sink, table etc) and remove the numbering on each line''')['content']        
    write_jsons(res, filename = "new_living_room_prompts.txt")
    time.sleep(2)
```

```python
def string_to_json_list(string):
    # Split the string into individual JSON objects
    json_objects = string.split("}")
    
    # Remove the last empty element
    json_objects.pop()
    
    # Convert each JSON object to a dictionary
    json_dicts = []
    for json_object in json_objects:
        json_dict = json.loads(json_object)
        json_dicts.append(json_dict)
    
    return json_dicts
```
